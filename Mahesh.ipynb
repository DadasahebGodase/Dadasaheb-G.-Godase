**<span style="color:blue">1. Opening a File**

> <span style="color:Brown">***Syntax------open(Filename, Mode)***
> 
> Filename = Any name that you want.
>
> Mode = Different modes for opening a file. 

The modes are as follows:
    
1. **"r"** - Default value. Opens file for reading only, error occurs if the file does not exists.

    
2. **"a"** - Append. Opens a file for appending, creates the file if it does not exists.

    
3. **"w"** - Write. Opens a file for writting, creats the file if it does not exists.

    
4. **"x"** - Create. Creats the specified file, returns an error if the file exists.

    
5. **"r+"** - Open the file for reading and writing.

    
6. **"w+"** - Open the file for reading and writing. For existing file, data is truncated and over-written.

    
7. **"a+"** - Open the file for reading and writing. The file is created if it does not exist. 

***
 
 

```python
#opening a file in "read" mode

a=open("F:/A/B.txt","r")
#a.close()
```

```python
b=open("F:/A/B.txt")
#b.close()
```

```python
#opening a file in "write" mode

c=open("F:/A/B.txt","w")
#c.close()
```

```python
# Creation of new file using "x" mode

a=open("F:/A/A.txt","x")
```

**<span style="color:blue">4. Tell and Seek functions**

* **Tell** - Where the current position of the file is?

        
* **Seek** - To get back to the start of the file (or anywhere else in the file), use the seek(int).
***

```python
d=open("F:/A/F.txt","r")
d.read()
#d.close()
```

```python
d.tell()  # I want to check where I am in the file d
```

```python
d.seek(2)  # I want to come at the second character
```

```python
d.tell()   # Now I want to check where I am in the file
```
